# COVID-19 Dataset

It is a small dataset on the covid-19 from all the different countries all around the world.
The data hold the record of the date 29th April 2020.

```python
import pandas as pd
```

```python
data = pd.read_csv('covid_19_data.csv')
```

```python
data
```

```python
data.count()
```

```python
data.isnull().sum()
```

## Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
data.head()
```

```python
data.groupby('Region').sum()
```

```python
data.groupby('Region')['Confirmed'].sum()
```

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False)
```

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=True)
```

```python
data.groupby('Region')['Confirmed', 'Recovered', 'Deaths'].sum()
```

## Remove all the records where the Confirmed Cases is Less Than 10.

```python
data.Confirmed<10
```

```python
data[data.Confirmed<10]
```

```python
data[~(data.Confirmed<10)]
```

## In which Region, maximum number of Confirmed cases were recorded ?

```python
data.groupby("Region").Confirmed.sum()
```
